```python
import wandb
api = wandb.Api()

project = "ntuyianchen/g_vz_eflow_2pcgnn_slurm"

def print_epoch_time(date):
    result = {}
    for run in api.runs(project):
        if date in run.id:
            run = api.run(project+"/"+run.id)
            if run.config["model_name"] != "Classical2PCGNN":
                print(run.id)
                history = run.history(keys=["epoch", "epoch_time"], x_axis="epoch")
                print(history)
                print("_"*100)

                model_structure = run.id.split()[0]
                average_epoch_time = history["epoch_time"].mean()
                result[model_structure] = average_epoch_time
    return result

result1 = print_epoch_time("20230822_082220")
result2 = print_epoch_time("20230822_082211")
result3 = print_epoch_time("20230822_082355")
result4 = print_epoch_time("20230822_082305")
```
